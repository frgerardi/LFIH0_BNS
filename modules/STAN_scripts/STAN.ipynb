## STAN scripts construction

```python
BNS_code_3D = """

data {
    int<lower=0> J; // number of sources
    vector[J] z_obs; // redshift for J sources
    vector<lower=0>[J] D_obs; // distance for J sources 
    real<lower=0> frac_err_D;
    vector[J] v_obs; // velocities for J sources
    real mean_v;
    real<lower=0> sigma_vlos;
    real<lower=0> sigma_z;
    real<lower=0> sigma_v;
    real<lower=0,upper=0.99> zmax;
    real<lower=0,upper=0.99> zmin;
}

transformed data {}

parameters {
    real<lower=50,upper=90> H;
    real<lower=-2,upper=1> q;
    vector<lower=zmin,upper=zmax>[J] z_cosmo_true;
    vector[J] v_true;
}

transformed parameters {

    vector<lower=0.0>[J] D_true;
    vector[J] z_tot_true;
    real G = 4.3*10^(-9); 
    real c = 299792;
    real N_bar_sel;


    for (j in 1:J){
        D_true[j] = c*z_cosmo_true[j]*(1.0+0.5*(1-q)*z_cosmo_true[j])/H; 
        z_tot_true[j] = z_cosmo_true[j] + (1+z_cosmo_true[j])*v_true[j]/c;   
    }  
    
}

model {

    H ~ normal(70.0, 20.0);
    q ~ normal(-0.55, 0.5);  
    
    target+= -3.0 * log(H) -log(1+z_cosmo_true)+2.0*log(z_cosmo_true)+log(1.0-2.0*(1+q)*z_cosmo_true)
              - log((zmax^2)/2 + log(1+zmax) - zmax - (zmin^2)/2 - log(1+zmin) + zmin \
                    - 2*(1+q)*((zmax^3)/3 - (zmax^2)/2 + zmax - log(1+zmax) -\
                                    (zmin^3)/3 + (zmin^2)/2 - zmin + log(1+zmin))) + 3.0 * log(H); 
    
//    target+= - J*log(N_bar_sel);


    v_true ~ normal(mean_v, sigma_vlos);
    
    z_obs ~ normal(z_tot_true, sigma_z); 
    v_obs ~ normal(v_true, sigma_v);

    for (j in 1:J){
        D_obs[j] ~ normal(D_true[j], D_true[j] * frac_err_D);
    } 

}


"""

sm_BNS_3D = pystan.StanModel(model_code=BNS_code_3D)
with open('3D_model.pkl', 'wb') as f:
    pkl.dump(sm_BNS_3D, f)
```

```python
BNS_code_3Dsel = """

functions{

    real N_numerical(real H, real q, vector c){
        
        return c[1] + c[2]*H + c[3]*q+
                c[4]*H*q + c[5]*H^2 + c[6]*q^2+
                c[7]*H^2*q + c[8]*q^2*H + c[9]*H^3 + c[10]*q^3 +\
                c[11]*H^4 + c[12]*q^4 + c[13]*q*H^3 + c[14]*H*q^3 + c[15]*(H*q)^2 ;
    }

}

data {
    int<lower=0> J; // number of sources
    vector[J] z_obs; // redshift for J sources
    vector<lower=0>[J] D_obs; // distance for J sources 
    real<lower=0> frac_err_D;
    vector[J] v_obs; // velocities for J sources
    real mean_v;
    real<lower=0> sigma_vlos;
    real<lower=0> sigma_z;
    real<lower=0> sigma_v;
    real<lower=0,upper=0.99> zmax;
    real<lower=0,upper=0.99> zmin;
    int<lower=0> N; // number of coefficients
    vector[N] numerical_coefficients;
}

transformed data {}

parameters {
    real<lower=50,upper=90> H;
    real<lower=-2,upper=1> q;
    vector<lower=zmin,upper=zmax>[J] z_cosmo_true;
    vector[J] v_true;
}

transformed parameters {

    vector<lower=0.0>[J] D_true;
    vector[J] z_tot_true;
    real G = 4.3*10^(-9); 
    real c = 299792;
    real N_bar_sel;


    for (j in 1:J){
        z_tot_true[j] = z_cosmo_true[j] + (1+z_cosmo_true[j])*v_true[j]/c;
        D_true[j] = c*z_cosmo_true[j]*(1.0+0.5*(1-q)*z_cosmo_true[j])/H;    
    }  
    


    N_bar_sel = N_numerical(H,q,numerical_coefficients);

}

model {

    H ~ normal(70.0, 20.0);
    q ~ normal(-0.55, 0.5);  
    
    target+= -3.0 * log(H) -log(1+z_cosmo_true)+2.0*log(z_cosmo_true)+log(1.0-2.0*(1+q)*z_cosmo_true);
    
    target+= - J*log(N_bar_sel);

    v_true ~ normal(mean_v, sigma_vlos);
    
    z_obs ~ normal(z_tot_true, sigma_z); 
    v_obs ~ normal(v_true, sigma_v);

    for (j in 1:J){
        D_obs[j] ~ normal(D_true[j], D_true[j] * frac_err_D);
    } 
  

}


"""

sm_BNS_3Dsel = pystan.StanModel(model_code=BNS_code_3Dsel)
with open('3Dsel_model.pkl', 'wb') as f:
    pkl.dump(sm_BNS_3Dsel, f)
```

```python
def poly_fit(XY,*c):
    H = XY[:,0]
    q = XY[:,1]
    return c[0] + c[1]*H + c[2]*q+\
            c[3]*H*q + c[4]*H**2 + c[5]*q**2+\
            c[6]*H**2*q + c[7]*q**2*H + c[8]*H**3 + c[9]*q**3 +\
            c[10] * H ** 4 + c[11] * q ** 4 + c[12] * q * H ** 3 + c[13] * H * q ** 3 + c[14] * (H * q) ** 2
```

```python
if new_fit == False:
    if selection == True:
        numerical_fit = np.loadtxt('N_bar_numerical_coefficients_3Dsel.txt')
    N_coeff = np.zeros((n_coeff_fit))
    for i in range (N_coeff.shape[0]):
        N_coeff[i] = numerical_fit[i]
else:
#grid construction
    H0_grid_bins = np.linspace (lower[0],upper[0],grid_bins)
    q0_grid_bins = np.linspace (lower[1],upper[1],grid_bins)
    X,Y = np.meshgrid(H0_grid_bins,q0_grid_bins)
    XY = np.array([X.ravel(), Y.ravel()]).T

#steps from paper section IV F to fit N(H0,q0) - other parameters fixed
    #1. model parameters (also in set_params)
    rate = 2.0e-4 
    obs_period = 1.0
    #2. selec
    z_max = z_range[1]
    z_min = z_range[0]
    #3. calculate N_bar_max
    N_sel = np.zeros((grid_bins*grid_bins))
    N_sims = np.zeros((grid_bins*grid_bins))
    for i in range (grid_bins*grid_bins):
        print(XY[i])
        z_integral = z_max**2/2 + np.log(1+z_max) - z_max \
                    - z_min**2/2 - np.log(1+z_min) + z_min \
                    - 2*(1+XY[i,1])*(z_max**3/3 - z_max**2/2 + z_max - np.log(1+z_max) -\
                                    z_min**3/3 + z_min**2/2 - z_min + np.log(1+z_min))
        #print(z_integral) 
        z_integral *= 4.0 * np.pi * (c / XY[i,0]) ** 3
        N_poisson = z_integral * rate * obs_period
        #4. draw N_max from Poisson of mean N_bar_max
        N = int(np.random.poisson(N_poisson))
        print('total number of sources = ',N)
        #5. sims of N_max sources
        sims = data_generation.data_simulator_for_N(XY[i,:],N)
        
        print('max distance', np.amax(sims[:,1]))
        print('max true distance', np.amax(sims[:,3]))
        print('max sigma D', np.amax(sims[:,4]))

        for k in range (N):
            if sims[k,2] >= 12:
                N_sel[i] += 1
            
        print('number of selected sources = ',N_sel[i])

        if check == True and (i==0):
            plt.scatter(sims[:,1],sims[:,2])
            indx = np.where(sims[:,2]>12)
            plt.scatter(sims[indx,1],sims[indx,2])
            plt.show()
            plt.close()
            plt.scatter(sims[:,0],sims[:,1])
            indx = np.where(sims[:,2]>12)
            plt.scatter(sims[indx,0],sims[indx,1])
            plt.show()           
            
#reallocate everything as in a grid
    N_sel_grid = np.zeros((grid_bins,grid_bins))
    j=0
    while j<grid_bins:
        for i in range (grid_bins): #move along H0
            N_sel_grid[j,i]=N_sel[i+grid_bins*j]
        j += 1
        
#fit
#scipy.optimize.curve_fit(f, xdata, ydata, p0=None)
#ydata = f(xdata, *params) + eps

    N_coeff, cov = so.curve_fit(poly_fit, XY, N_sel, p0=np.zeros(n_coeff_fit))
    #N_coeff : Optimal values for the parameters so that the sum of the squared residuals of 
    #f(xdata, *popt) - ydata is minimized.
    
    n_bar_fit = np.zeros((grid_bins, grid_bins))
    for i in range(grid_bins):
        for j in range(grid_bins):
            coord = np.zeros((1,2))
            coord[0,0] = H0_grid_bins[i]
            coord[0,1] = q0_grid_bins[j]
            n_bar_fit[i, j] = poly_fit(coord, *N_coeff)
    
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.contour3D(X, Y, n_bar_fit, 50,cmap='binary')
    ax.contour3D(X, Y, N_sel_grid, 50)
    ax.set_xlabel('H0')
    ax.set_ylabel('q0')
    ax.set_zlabel('z')
    
    #normalizing N_bar in such a way to recover the number of observed BNS (n_sources) at fiducial cosmology
    fiducial_params = np.zeros((1,2))
    fiducial_params[:] = theta_fiducial
    norm_factor = poly_fit(fiducial_params, *N_coeff)
    print(norm_factor)
    N_coeff *= float(n_sources) / norm_factor
    
    if selection == True:
        np.savetxt('N_bar_numerical_coefficients_3Dsel.txt',N_coeff)
```
